```python
def add_noise(arr):
    arr1 = arr.copy()
    ver = np.zeros((1, 100))
    ver = ver[0]
    for i in range(12):
        ver[i] = 1
    for i in range(arr1.shape[0]):
        for j in range(arr1.shape[1]):
            b = random.choice(ver)
            if b != 0:
                arr1[i, j] = np.array([255, 255, 255])
    return arr1
```

```python
img2 = add_noise(img)
```

```python
plt.imshow(img2)
```

```python
def bubble_sort(arr):
    arr1 = arr.copy()
    for j in range(len(arr1)):
        for i in range(len(arr1)):
            try:
                if sum(arr1[i+1]) < sum(arr1[i]):
                    arr1[i], arr1[i+1] = arr1[i+1], arr1[i]
            except:
                try:
                    if arr1[i+1] < arr1[i]:
                        arr1[i], arr1[i+1] = arr1[i+1], arr1[i]
                except:
                    pass
    return arr1
```

```python
x = [1, 3, 2, 4]
```

```python
bubble_sort(x)
```

```python
def rang_filt(arr):
    arr1 = arr.copy()
    arr1 = arr1.reshape(arr1.shape[0], arr1.shape[1]*3, 1)
    for i in range(1, arr1.shape[0]-1):
        print(i)
        for j in range(1, arr1.shape[1]-1):
            square = [arr1[i, j], arr1[i+1, j], arr1[i-1, j], arr1[i, j+1], arr1[i, j-1]]
            square = bubble_sort(square)
            arr1[i, j] = square[0]
            arr1[i+1, j] = square[1]
            arr1[i-1, j] = square[2]
            arr1[i, j+1] = square[3]
            arr1[i, j-1] = square[4]
    arr1 = arr1.reshape(arr1.shape[0], arr1.shape[1]/3, 3)
    return arr1
```
